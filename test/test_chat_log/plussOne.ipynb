add +1 hour to the text file

```python
import re
from datetime import datetime, timedelta
```

```python
input_filename = '[02-12-26].txt'
output_filename = '[02-12-26]_fixed.txt'

# 1. Open the source file and the new destination file
with open(input_filename, 'r', encoding='utf-8') as infile, \
     open(output_filename, 'w', encoding='utf-8') as outfile:

    for line in infile:
        # 2. Search for the pattern [YYYY-MM-DD HH:MM:SS UTC]
        # We capture the date/time part inside the parentheses group
        match = re.search(r'\[(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) UTC\]', line)

        if match:
            # Extract the string found (e.g., "2026-02-12 19:47:25")
            original_time_str = match.group(1)
            
            # 3. Convert string to a datetime object
            time_obj = datetime.strptime(original_time_str, '%Y-%m-%d %H:%M:%S')
            
            # 4. Add 1 hour
            new_time_obj = time_obj + timedelta(hours=1)
            
            # 5. Convert back to string
            new_time_str = new_time_obj.strftime('%Y-%m-%d %H:%M:%S')
            
            # 6. Replace the old time with the new time in the line
            # We replace the specific captured part
            new_line = line.replace(original_time_str, new_time_str)
            
            outfile.write(new_line)
        else:
            # If a line doesn't match the format (e.g. empty lines), write it unchanged
            outfile.write(line)

print(f"Done! Processed lines saved to {output_filename}")


```
