```python
import pandas as pd
import re
from collections import Counter
```

```python
# Read the list of filenames from the configuration file
with open('../file_list.txt', 'r', encoding='utf-8') as config_file:
    file_names = config_file.read().splitlines()

# Regex pattern to match the data format
pattern = r'\[(.*?)\] (.*?): (.*)'

# Initialize an empty list to store parsed data
datalist = []
stream_count = 0
# Iterate over each specified file
for file in file_names:
    full_path = f"../data/{file}"
    with open(full_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            match = re.match(pattern, line)
            if match:
                date, user, message = match.groups()
                datalist.append([date, user, message,stream_count])
    stream_count = stream_count + 1

# Create a DataFrame from the parsed data
data = pd.DataFrame(datalist, columns=["date", "user", "message","stream"])
data['date'] = pd.to_datetime(data['date'])

# Create a DataFrame from the parsed data
data = pd.DataFrame(datalist, columns=["date", "user", "message","stream"])
```

```python
# Assume 'data' DataFrame and 'merged_list' are predefined.
# Ensure 'data' has 'message', 'date', 'user', 'stream' columns.
# Ensure 'data['date']' is already in datetime format.

# ... (Steps 1-3 for preprocessing merged_list and creating capturing_regex are the same) ...

# 1. Preprocess merged_list:
if 'merged_list' not in locals() or not isinstance(merged_list, (list, pd.Series, set, tuple)):
    unique_search_terms = []
else:
    processed_terms = []
    for term in merged_list:
        if isinstance(term, str) and term.strip():
            processed_terms.append(term)
    unique_search_terms = list(set(processed_terms))

# 2. Handle cases where there are no valid search terms
if not unique_search_terms:
    word_stats = pd.DataFrame(columns=['Word', 'Last Used', 'Total Count', 'Unique Users', 'Unique Streams'])
else:
    # 3. Construct the regex pattern
    regex_parts = [rf'\b{re.escape(term)}\b' for term in unique_search_terms]
    capturing_regex = f"({'|'.join(regex_parts)})"

    # 4. Extract all occurrences
    all_matches = data['message'].str.extractall(capturing_regex)

    if all_matches.empty:
        word_stats = pd.DataFrame(columns=['Word', 'Last Used', 'Total Count', 'Unique Users', 'Unique Streams'])
    else:
        # 5. Rename the extracted word column
        all_matches = all_matches.rename(columns={0: 'MatchedWord'})

        # --- MODIFIED SECTION START ---

        # 6. Prepare 'all_matches' for merging by resetting its index.
        #    This turns the MultiIndex levels (like the original message index and match number)
        #    into regular columns.
        all_matches_flat = all_matches.reset_index()

        #    Determine the name of the column in 'all_matches_flat' that holds the original data's index.
        #    This column originates from the first level of 'all_matches's MultiIndex.
        #    The name of this level is inherited from data.index.name.
        #    If data.index was unnamed, reset_index() names this column 'level_0'.
        original_data_index_col_name_in_flat = all_matches.index.names[0]
        if original_data_index_col_name_in_flat is None:
            original_data_index_col_name_in_flat = 'level_0'

        #    For clarity and consistency, rename this column to 'original_message_index'.
        #    Then select only this renamed column and 'MatchedWord' for the merge.
        matches_for_merge = all_matches_flat.rename(
            columns={original_data_index_col_name_in_flat: 'original_message_index'}
        )[['original_message_index', 'MatchedWord']]

        #    Perform the merge:
        #    - Left DataFrame: The relevant columns from the original 'data'.
        #    - Right DataFrame: The prepared 'matches_for_merge'.
        #    - `left_index=True`: Use the index of `data[['date', 'user', 'stream']]` for the join.
        #    - `right_on='original_message_index'`: Join on the 'original_message_index' column
        #                                           from `matches_for_merge`.
        #    - `how='right'`: Ensures all matched words are included, bringing in their
        #                     corresponding data from the left DataFrame. This preserves all matches.
        df_for_aggregation = pd.merge(
            left=data[['date', 'user', 'stream']],
            right=matches_for_merge,
            left_index=True,
            right_on='original_message_index',
            how='right'
        )
        # df_for_aggregation now has 'date', 'user', 'stream', 'original_message_index', 'MatchedWord'
        # as columns, and a simple RangeIndex.

        # Step 7 (Resetting index of df_for_aggregation and renaming 'level_0')
        # is no longer needed as 'original_message_index' is already a column with the correct name.

        # 8. Group by the 'MatchedWord' and aggregate statistics:
        #    'original_message_index' column is now directly available for 'nunique'.
        word_stats_aggregated = df_for_aggregation.groupby('MatchedWord').agg(
            Last_Used=('date', 'max'),
            Total_Count=('original_message_index', 'nunique'), # Counts unique messages
            Unique_Users=('user', 'nunique'),
            Unique_Streams=('stream', 'nunique')
        ).reset_index()

        # --- MODIFIED SECTION END ---

        # 9. Rename columns to the final desired format (this step remains the same)
        word_stats = word_stats_aggregated.rename(columns={
            'MatchedWord': 'Word',
            'Last_Used': 'Last Used',
            'Total_Count': 'Total Count',
            'Unique_Users': 'Unique Users',
            'Unique_Streams': 'Unique Streams'
        })
word_stats.head()
```
