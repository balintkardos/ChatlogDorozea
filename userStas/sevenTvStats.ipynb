```python
# Read the list of filenames from the configuration file
with open('../file_list.txt', 'r', encoding='utf-8') as config_file:
    file_names = config_file.read().splitlines()

# Regex pattern to match the data format
pattern = r'\[(.*?)\] (.*?): (.*)'

# Initialize an empty list to store parsed data
datalist = []
stream_count = 0
# Iterate over each specified file
for file in file_names:
    full_path = f"../data/{file}"
    with open(full_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            match = re.match(pattern, line)
            if match:
                date, user, message = match.groups()
                datalist.append([date, user, message,stream_count])
    stream_count = stream_count + 1

# Create a DataFrame from the parsed data
data = pd.DataFrame(datalist, columns=["date", "user", "message","stream"])
data['date'] = pd.to_datetime(data['date'])

```

```python
# Get all unique usernames
unique_users = data['user'].unique()

# Create a mapping from lowercase username to all variants

user_variants = defaultdict(set)
for user in unique_users:
    user_variants[user.lower()].add(user)

# Find usernames with different capitalization
duplicate_users = {k: v for k, v in user_variants.items() if len(v) > 1}
```

```python
# Create a mapping from all variants to the canonical (sorted first) variant
variant_map = {}
for variants in duplicate_users.values():
    sorted_variants = sorted(variants)
    canonical = sorted_variants[0]
    for v in variants:
        variant_map[v] = canonical

# Replace usernames in 'user' column
data['user'] = data['user'].apply(lambda u: variant_map.get(u, u))
```

```python
# 1. Convert your emote list to a set for ultra-fast matching
emote_set = set(final_emote_list)

# 2. Function to split messages by space and extract valid 7tv emotes
def get_valid_emotes(text):
    if not isinstance(text, str):
        return []
    return [word for word in text.split() if word in emote_set]

# 3. Apply extraction and drop messages that don't contain any target emotes
data['extracted_emotes'] = data['message'].apply(get_valid_emotes)
exploded = data.explode('extracted_emotes').dropna(subset=['extracted_emotes'])

# 4. Group by stream and emote to create a clean frequency matrix
# Rows = Streams, Columns = Emotes, Values = Counts
counts_matrix = exploded.groupby(['stream', 'extracted_emotes']).size().unstack(fill_value=0)

# 5. Ensure your matrix is sorted chronologically by your streams
# (If your stream IDs are dates or sequential strings, sort the index)
counts_matrix = counts_matrix.sort_index()
```

```python
# Divide each row by its total emote count to get the share of chat per stream
relative_matrix = counts_matrix.div(counts_matrix.sum(axis=1), axis=0)
```

```python
# Split data using index slicing
historical_baseline = relative_matrix.iloc[:-25]
recent_25_streams = relative_matrix.iloc[-25:]

# Calculate the baseline historical average and standard deviation for each emote
historical_mean = historical_baseline.mean()
historical_std = historical_baseline.std().replace(0, 1) # Prevents division by zero errors

# Calculate Z-scores across all of the last 25 streams
z_scores = (recent_25_streams - historical_mean) / historical_std

# Find the highest peak Z-score achieved by any emote in this window
top_trending_peaks = z_scores.max().sort_values(ascending=False)

print("Top Peak Trending Emotes in the Last 25 Streams:")
print(top_trending_peaks.head(10))
```

```python
# Re-extract raw counts for the 25 window to set a minimum threshold filter
total_counts_in_window = counts_matrix.iloc[-25:].sum()

# Split the 25-stream relative matrix into older (20) vs newest (5)
long_window_avg = recent_25_streams.iloc[:20].mean()
short_window_avg = recent_25_streams.iloc[20:].mean()

# Calculate relative growth velocity
momentum = (short_window_avg - long_window_avg) / long_window_avg

# Apply a filter (e.g., must have appeared at least 100 times total to filter out noise)
trending_momentum = momentum[total_counts_in_window > 100].sort_values(ascending=False)

print("\nTop Sustained Growth Emotes (Momentum):")
print(trending_momentum.head(10))
```

```python
# Combine total usage and unique usage into one dataframe
user_stats = exploded.groupby('user')['extracted_emotes'].agg(['size', 'nunique'])
user_stats.columns = ['total_instances', 'unique_emotes']

# Calculate a variety ratio (Percentage of their emote usage that is unique)
user_stats['variety_ratio'] = user_stats['unique_emotes'] / user_stats['total_instances']

# Filter out users who have only typed a few emotes total to avoid 1/1 = 100% distortion
true_collectors = user_stats[user_stats['total_instances'] > 50].sort_values(by='unique_emotes', ascending=False)

print("\n--- TOP EMOTE COLLECTORS (MIN 50 EMOTES TYPED) ---")
print(true_collectors.head(10).reset_index())
```

```python
# Assuming 'exploded' already exists from your extraction step:
# exploded = data.explode('extracted_emotes').dropna(subset=['extracted_emotes'])

# 1. Count how many times each user used each specific emote
emote_user_counts = (
    exploded.groupby(['extracted_emotes', 'user'])
    .size()
    .reset_index(name='user_count')
)

# 2. Count the absolute total usage of each emote across the entire community
emote_total_counts = (
    exploded.groupby('extracted_emotes')
    .size()
    .reset_index(name='total_count')
)

# 3. Merge the datasets to compare individual usage against total community usage
merged_emotes = pd.merge(emote_user_counts, emote_total_counts, on='extracted_emotes')

# 4. Calculate what percentage of total usage came from this single user
merged_emotes['usage_percentage'] = (merged_emotes['user_count'] / merged_emotes['total_count']) * 100

# 5. Filter for emotes where a single user accounts for 50% or more of the usage.
# Change 'total_count >= 10' to a higher number if you want to find even more dedicated spammers.
personal_emotes_df = merged_emotes[
    (merged_emotes['usage_percentage'] >= 30.0) & 
    (merged_emotes['total_count'] >= 10)
].sort_values(by='total_count', ascending=False)

# Clean up layout for printing
personal_emotes_df = personal_emotes_df[[
    'extracted_emotes', 'user', 'user_count', 'total_count', 'usage_percentage'
]].rename(columns={'extracted_emotes': 'emote'})

print("--- DETECTED PERSONAL EMOTES (>= 50% OWNERSHIP) ---")
print(personal_emotes_df.head(20).to_string(index=False))
```
